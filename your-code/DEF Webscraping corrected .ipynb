#### List all language names and number of related articles in the order they appear in wikipedia.org

````python
# This is the url you will scrape in this exercise
url_9 = 'https://www.wikipedia.org/'
html = requests.get(url_9).content
soup9 = BeautifulSoup(html, "html.parser")
````

````python
#your code: language names and number of related articles

wiki_table = soup9.find_all('div',{'class':'central-featured'})
wiki_table= [row.text.strip().split("\n") for row in wiki_table]
wiki_table= wiki_table[0] # shows [0]= country, wuth steps of 7 [1]= count 

#create a list op countries and a list of numbers 
countries = wiki_table[0::7] 
numbers = wiki_table[1::7] 
numbers_list=[]

# cleaning the numbers list 
for x in range(len(numbers)): 
    pattern= '[0-9]+'
    number = re.findall(pattern, numbers[x])
    number = ("".join(number))
    numbers_list.append(number)

# concerting the list in a dataframe     
wiki_names_numbers = pd.DataFrame(list(zip(countries,numbers_list)),
                                    columns =['countries','numbers_list'])
wiki_names_numbers
````
