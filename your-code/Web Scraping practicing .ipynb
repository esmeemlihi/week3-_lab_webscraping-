```python
import requests

url = 'https://www.reuters.com/article/us-shazam-m-a-apple-eu/eu-clears-apples-purchase-of-shazam-idUSKCN1LM1TZ'
html = requests.get(url).content
html[0:600]
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(html, "html.parser")
#soup = BeautifulSoup(html, "html5lib")
#soup = BeautifulSoup(html, "lxml")
soup
```

```python
tags = ['h1', 'h2', 'h3', 'h4', 'h5', 'h6', 'h7', 'p']
text = [element.text for element in soup.find_all(tags)] 
# this list comprehension loops to al the tags and plots the textstrings 
text
```

```python
print('\n'.join(text))
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_European_countries_by_life_expectancy'
html = requests.get(url).content
soup2 = BeautifulSoup(html, "html.parser")
```

```python
table = soup2.find_all('table',{'class':'sortable wikitable'})[0]
table
```

```python
rows = table.find_all('tr') # find als the contect that are marhet in html with <tr>...</tr>
rows = [row.text.strip().split("\n") for row in rows] 
#strip() takes away all the extra spaces # and it plots the text in seperated rows via "\n"
rows2 = [[x for x in row if x] for row in rows] # if x means it will only sellect filled rows not empty ones 

rows2
```

```python
import pandas as pd
colnames = rows2[0]
data = rows2[1:]

# the \xa0 is a code (first wrapped) so it wass plotted firts and now filtered out via panda 
df = pd.DataFrame(data, columns=colnames)
df
```
